```python
#1 PRINT FIRST 10 NATURAL NUMBERS
def first_10_naturalnumb():
  for n in range (1,11):
    print(n)
first_10_naturalnumb()

```

```python
#2 SUM OF FIRST N NATURAL NUMBERS

def sum_n(n):
  sum=0
  for i in range(1,n+1):
    sum= sum+i
  print(sum)
sum_n(3)

```

```python
#3 REVERSE A NUMBER
n=int(input("enter a number: "))
def reverse(n):
  rev =0
  while n>0:
    digit= n % 10
    rev = rev*10 + digit
    n=n//10
  return rev

print(reverse(n))
```

```python
#4 COUNT DIGITS IN A NUMBER
def count_digits(n):
  count=len(n)
  return count
n=input("enter a number : ")
print(count_digits(n))
```

```python
#5 CHECK PALINDROME NUMBER
def check_palindrome(x):
  if (str(x)==str(x[::-1])):
    print("PALINDROME")
  else:
    print("NOT A PALINDROME")

a= input("enter a word: ")
check_palindrome(a)

```

```python
#6 GENERATE FIBONACCI SERIES
def fibonacci_series(n):
  a,b=0,1
  print("FIBONACCI SERIES: ")
  for i in range (n):
    print(a,end=" ")
    a,b= b, a+b
n=int(input("enter the number of terms : "))
fibonacci_series(n)


```

```python
#7 CALCULATOR
def calculate(a,b,operator):
  if (operator=="add"):
    return a+b
  elif operator=="subtract":
    return a-b
  elif operator=="multiply":
    return a*b
  elif operator=="divide":
    return a/b
print(calculate(2,4,"add"))
```
